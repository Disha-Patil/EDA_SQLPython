```python
#importing pandas library
import pandas as pd
```

```python
#reading the csv file
df=pd.read_csv('bank_transactions.csv')
```

```python
#renaming column CustomerDOB to DOB
df=df.rename(columns={'CustomerDOB': 'DOB'})
```

```python
#renaming column CustGender to Gender
df=df.rename(columns={'CustGender': 'Gender'})
```

```python
#renaming column CustLocation to Location
df=df.rename(columns={'CustLocation': 'Location'})
```

```python
#renaming column CustAccountBalance to AccountBalance
df=df.rename(columns={'CustAccountBalance': 'AccountBalance'})
```

```python
#renaming column TransactionAmount(INR) to TransactionAmount
df=df.rename(columns={'TransactionAmount (INR)': 'TransactionAmount'})
```

```python
#representing the table df
df
```

```python
#importing datetime library
from datetime import datetime
```

```python
#assigning today's date to x
x=datetime.today()
```

```python
#declaring an empty list age
age=[]
```

```python
#converting the column DOB in its formatted form yyyy-mm-dd
df.DOB = pd.to_datetime(df.DOB, format='%d/%m/%Y')
```

```python
#representing the column DOB in its formatted form
df.DOB
```

```python
#retrieving the year of DOB
year=df.DOB.dt.year
```

```python
#retrieving the month of DOB
month=df.DOB.dt.month
```

```python
#retrieving the year of present date
y=int(x.strftime('%Y'))
```

```python
#retrieving the month of present date
m=int(x.strftime('%m'))
```

```python
#finding the age and adding it to the list age
for i in range(0,len(year)):
    if(month[i]>m and (month[i]-m)<6):
        a=y-year[i]
    elif(month[i]<m and (m-month[i])<6):
        a=y-year[i]
    elif(month[i]<m and (m-month[i])>=6):
        a=y-year[i]+1
    else:
        a=y-year[i]+1
    age.append(a)
```

```python
#representing the list age
age
```

```python
#declaring variable sum as 0
sum=0
```

```python
#finding sum of all positive ages
for i in range(0, len(age)):
    if(age[i]>0):
        sum=sum+age[i]
```

```python
#finding average of all positive ages
avg=sum/len(age)
```

```python
#representing average in decimal form
avg
```

```python
#rounding off average age to its nearest integer
d=int(avg)
```

```python
#representing average age to its nearest integer
d
```

```python
#assigning average age to all negative ages
for i in range(0, len(age)):
    if(age[i]<0):
        age[i]=d
```

```python
#making column age into the table
df['age']=age
```

```python
#representing column age
df.age
```

```python
#finding all unique values of age
e=df.age.unique()
```

```python
#representing all unique values of age
e
```

```python
#sorting age
e.sort()
```

```python
#assigning the sorted age to a variable v
v=sorted(e)
```

```python
#representing the variable v
v
```

```python
#appending top ten values of ages into a newly made list k
k=[]
for i in range (0,9):
    r=v[len(v)-2-i]
    k.append(r)
```

```python
#representing that list
k
```

```python
#representing customer ID, transaction amount, account balance and age 
for i in k:
    for j in range(0, len(age)):
        if(df.age[j]==i):
            print(df.CustomerID[j], end=' ')
            print(df.TransactionAmount[j], end=' ')
            print(df.AccountBalance[j], end=' ')
            print(df.age[j])
```
