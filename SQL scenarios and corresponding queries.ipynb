```python
#Importing SQLite3 and connecting to database
import sqlite3
conn = sqlite3.connect('transactiondata.db')
```

```python
#Creating cursor
cur = conn.cursor()
```

```python
#Scenario 1: Find the cumulative Transaction Amount for each month, across all customers, locations, etc. 
cur.execute("SELECT SUM(TransactionAmount) FROM Transactions WHERE TransactionDate LIKE '%/8/%'")#Finding cumulative transaction amount for August
```

```python
rows=cur.fetchall()
```

```python
for x in rows:
    print(x)
```

```python
#Scenario 2: Find the cumulative Account Balance for each month, across all customers, locations, etc.
cur.execute("Select substr(TransactionDate, (INSTR(TransactionDate, '/')+1)), SUM(TransactionAmount) from (SELECT *, substr(TransactionDate, (INSTR(TransactionDate, '/')+1)) FROM Transactions WHERE TransactionDate!='TransactionDate')  group by substr(TransactionDate, (INSTR(TransactionDate, '/')+1))")
```

```python
rows=cur.fetchall()
```

```python
for x in rows:
    print(x)
```

```python
#Scenario 3: Find the cumulative transaction amount for each location.
cur.execute("SELECT CustLocation, SUM(TransactionAmount) FROM Transactions WHERE CustLocation!='CustLocation' GROUP BY(CustLocation)")
#(Note: The above query has the condition WHERE CustLocation!='CustLocation' because, for some reason, the first row of the SQL table is the column headers). 
```

```python
rows=cur.fetchall()
```

```python
for x in rows:
    print(x)
```

```python
#Scenario 5: Find the Top 10 customers whose total transaction amount is the highest
cur.execute("SELECT  CustomerID, SUM(TransactionAmount) FROM Transactions GROUP BY(CustomerID) ORDER BY SUM(TransactionAmount) DESC LIMIT 10")
```

```python
#Scenario 7: Find the Top 10 customers who have highest /lowest values of account balance (there maybe more than 1 row for a customer, in which case the row with the highest/lowest account balance is to be taken)
cur.execute("SELECT CustomerID, max(CustAccountbalance) AS maxbal from Transactions WHERE CustomerID!='CustomerID'  GROUP BY CustomerID ORDER BY maxbal DESC LIMIT 10")#Finding customers with highest account balance
#(Note: The above query has the condition WHERE CustomerID!='CustomerID' because, for some reason, the first row of the SQL table is the column headers).
```

```python
rows=cur.fetchall()
```

```python
for x in rows:
    print(x)
```

```python
cur.execute("SELECT CustomerID, min(CustAccountbalance) AS minbal from Transactions GROUP BY CustomerID ORDER BY minbal ASC LIMIT 10")#Finding customers with lowest account balance
```
